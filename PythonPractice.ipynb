```python
"""
ENHANCED VERSION: Rule-Based Chatbot with Personality & Nested Conditions
Demonstrates: Unique personality + nested conditions + expanded vocabulary
"""

import random
import re

class DecodeBot:
    """A rule-based chatbot with unique personality and nested logic"""

    def __init__(self):
        # Knowledge base with pattern matching (expanded vocabulary)
        self.responses = {
            'greeting': {
                'patterns': [r'\b(hello|hi|hey|sup|howdy)\b'],
                'responses': [
                    "Greetings, future AI engineer! 🚀",
                    "Hey there! Ready to write some deterministic logic?",
                    "Hello! Remember: Rules before probabilities!"
                ]
            },
            'feeling': {
                'patterns': [r'\b(happy|great|awesome|excited)\b'],
                'responses': [
                    "That's fantastic energy! That's exactly what we need to build AI.",
                    "Love the enthusiasm! Keep that momentum going.",
                    "Awesome! Positive vibes produce clean code."
                ]
            },
            'sad': {
                'patterns': [r'\b(sad|tired|stuck|confused)\b'],
                'responses': [
                    "Every engineer hits roadblocks. Take a break, then try again.",
                    "You've got this! Start small, build step by step.",
                    "Feeling stuck is part of learning. Want to review the basics?"
                ]
            },
            'identity': {
                'patterns': [r'\b(who are you|your name|what are you)\b'],
                'responses': [
                    "I'm DecodeBot - your rule-based mentor. Pure logic, zero hallucinations."
                ]
            }
        }

        self.fallback_responses = [
            "Hmm, I don't have a rule for that yet. Want to try 'help'?",
            "My deterministic brain doesn't know that one. Ask me something from my intent list!",
            "That's outside my rule set. Type 'help' to see what I understand."
        ]

        self.user_name = None

    def sanitize(self, text):
        """Input sanitization: lowercase, strip, remove extra spaces"""
        return ' '.join(text.lower().strip().split())

    def get_response_with_nested_logic(self, user_input):
        """
        Nested conditions for smarter responses
        Demonstrates control flow with multiple decision layers
        """
        sanitized = self.sanitize(user_input)

        # Layer 1: Check if user asked for name setting
        if 'my name is' in sanitized:
            name_match = re.search(r'my name is (\w+)', sanitized)
            if name_match:
                self.user_name = name_match.group(1).capitalize()
                return f"Nice to meet you, {self.user_name}! Let's build some AI logic together."
            else:
                return "I didn't catch that. Just say 'my name is [your name]'"

        # Layer 2: If we know user's name, use it for personalized responses
        if self.user_name and ('how are you' in sanitized or 'how do you do' in sanitized):
            return f"I'm functioning perfectly, {self.user_name}! Always ready to process if-else logic. How can I help?"

        # Layer 3: Pattern matching across knowledge base
        for intent, data in self.responses.items():
            for pattern in data['patterns']:
                if re.search(pattern, sanitized):
                    return random.choice(data['responses'])

        # Layer 4: Default fallback
        return random.choice(self.fallback_responses)

    def run(self):
        """Main execution loop with personality"""
        print("\n" + "="*60)
        print("🤖 DECODEBOT - THE PERSONALITY EDITION 🤖")
        print("="*60)
        print("Rule-Based | White Box | Deterministic Logic")
        print("Features: Pattern matching + Nested logic + Memory")
        print("-"*60)
        print("Tell me your name! ('my name is X')")
        print("Type 'bye', 'exit', or 'quit' to leave")
        print("="*60 + "\n")

        while True:
            user_input = input("You: ").strip()

            if not user_input:
                continue

            if user_input.lower() in ['bye', 'exit', 'quit']:
                if self.user_name:
                    print(f"Bot: Goodbye, {self.user_name}! Keep coding. Master the rules before the neural nets.\n")
                else:
                    print("Bot: Goodbye! Remember: Every master AI engineer started with if-else.\n")
                break

            response = self.get_response_with_nested_logic(user_input)
            print(f"Bot: {response}\n")

# Run the enhanced version
if __name__ == "__main__":
    import sys

    print("\nChoose version:")
    print("1. Standard (Hash Map + .get() method)")
    print("2. Enhanced (Personality + Nested Logic + Pattern Matching)")

    choice = input("Enter 1 or 2: ").strip()

    if choice == '2':
        bot = DecodeBot()
        bot.run()
    else:
        run_chatbot()  # From first implementation
```
